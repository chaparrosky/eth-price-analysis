```python
import pandas as pd

# Load the data
eth_data = pd.read_csv('eth-historical-data.csv', index_col='Date', parse_dates=True)

# Inspect the first few rows
print(eth_data.head())
```

```python
# Get a summary of the dataset
print(eth_data.describe())
```

```python
# Check for missing values
print(eth_data.isnull().sum())
```

```python
# Calculate daily returns
eth_data['Daily_Return'] = eth_data['Close'].pct_change()

# Calculate a 7-day moving average
eth_data['7-Day MA'] = eth_data['Close'].rolling(window=7).mean()

# Calculate a 30-day moving average
eth_data['30-Day MA'] = eth_data['Close'].rolling(window=30).mean()

# Drop any rows with NaN values that resulted from moving average calculations
eth_data.dropna(inplace=True)

# Display the first few rows to confirm the new features
print(eth_data.head())
```

```python
import matplotlib.pyplot as plt

# Plot the Close price and moving averages
plt.figure(figsize=(14, 7))
plt.plot(eth_data['Close'], label='Close Price')
plt.plot(eth_data['7-Day MA'], label='7-Day Moving Average')
plt.plot(eth_data['30-Day MA'], label='30-Day Moving Average')
plt.title('ETH Close Price and Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()

## From the graph, we can observe a general upward trend in Ethereum's price, especially noticeable around certain periods.
## The 7-day moving average responds more quickly to price changes compared to the 30-day moving average.
## Which is smoother and shows more general trends. These moving averages can help in identifying potential buy or sell signals.
```

```python
# Plot the daily returns
plt.figure(figsize=(14, 7))
plt.plot(eth_data['Daily_Return'], label='Daily Returns')
plt.title('ETH Daily Returns')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.legend()
plt.show()
```

```python
from datetime import timedelta

# Define the end date of your dataset (this would be the last date in eth_data)
end_date = eth_data.index[-1]

# Define time periods (3 months, 6 months, 9 months, 1 year)
periods = {
    '3 months': end_date - timedelta(days=90),
    '6 months': end_date - timedelta(days=180),
    '9 months': end_date - timedelta(days=270),
    '1 year': end_date - timedelta(days=365)
}

# Initialize a dictionary to store the best return days for each period
best_return_days = {}

# Loop through each period and find the day with the maximum return
for period, start_date in periods.items():
    # Filter the data for the given period
    period_data = eth_data.loc[start_date:end_date]
    
    # Find the day with the highest daily return in the period
    best_day = period_data['Daily_Return'].idxmax()
    best_return = period_data['Daily_Return'].max()
    
    # Get the buying price (Close price of the previous day)
    buy_price = eth_data.loc[best_day - timedelta(days=1), 'Close']
    
    # Get the selling price (Close price on the best return day)
    sell_price = eth_data.loc[best_day, 'Close']
    
    # Store the result in the dictionary
    best_return_days[period] = (best_day, buy_price, sell_price, best_return)

# Display the results
for period, (day, buy_price, sell_price, return_value) in best_return_days.items():
    print(f"In the last {period}:")
    print(f"   - Best return day: {day.date()}")
    print(f"   - Buying price (previous day close): ${buy_price:.2f}")
    print(f"   - Selling price: ${sell_price:.2f}")
    print(f"   - Return: {return_value:.2%}")
    print("")
```
